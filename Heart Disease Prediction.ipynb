Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
#loading the csv data to a Panda DataFrame
heart_data = pd.read_csv('/content/data.csv')
```

```python
#the number of rows and columns in datset
heart_data.shape
```

```python
#checking for null values
heart_data.isnull().sum()
```

```python
heart_data.describe()
```

```python
#checking the distribution of Target variable
heart_data['target'].value_counts()
```

1->Defective heart
0-> healthy heart

splitting the features and target

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

splitting the data into training data & test data

```python
X_train, X_test, Y_train, Y_test =train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

Logistic Regression Model

```python
model = LogisticRegression()
```

```python
#we are training LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

we are evaluating our model

Accuracy Score

```python
 X_train_prediction = model.predict(X_train)
 training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on  traininng data:',training_data_accuracy)
```

```python
 X_test_prediction = model.predict(X_test)
 test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on  test data:',test_data_accuracy)
```

Building a predictive system

```python
input_data=(61,1,0,148,203,0,1,161,0,0,2,1,3)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print('The person do not have a heart Disease')
else:
  print('the person has Heart Disease')
```
